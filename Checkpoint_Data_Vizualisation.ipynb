```python
# fill missing age with median age 
data["Age"].fillna(data["Age"].median(), inplace=True)
data["Cabin"].fillna("NULL", inplace=True)
data["Embarked"].fillna("NULL", inplace=True)
data
```

```python
def dist(feature,title):
    survived = data[data['Survived']=='Yes'][feature].value_counts()
    dead = data[data['Survived']=='No'][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['Survived','Dead']
    df.plot(kind='bar',stacked=True)
    plt.title(title)
```

```python
dist('Sex','Sex')
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    
plot_correlation_map( data )
```

```python
#The paragraph:
#The correlation map above shows the correlation coefficients between several variables or features.It is perfectly symmetrical.
#we notice that the correlation between “Pclass” and “Fare” is -0.55, which indicates 
#that they’re strongly negatively correlated. A higher class is strongly related to less Fare.
#The correlation between “Siblings and spouces” and “Parents and children” is 0.41, which indicates that they’re
#strongly positively correlated. More siblings and spouces is associated with more parents and children.
#The correlation between "SibSp" and "Pclass" is 0.083, which indicates that they’re basically not correlated.
#Also we notice that the correlation coefficients along the diagonal of the table are all equal to 1 because each 
#variable is perfectly correlated with itself.
#conclusion: It would be very difficult to understand the relationship between each variable by simply staring at the raw data. 
#A correlation map can help us quickly understand the correlations between each pair of variables.
```
